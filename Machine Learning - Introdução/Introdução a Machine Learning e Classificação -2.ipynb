```
import pandas as pd
```

```
uri = 'https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'
df = pd.read_csv(uri)

df.head()
```

```
map = {
    'home':'principal',
    'how_it_works':'como_funciona',
    'contact':'contato',
    'bought':'comprado'
}

df.rename(columns = map, inplace = True)
```

```
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```
df.shape
```

```
from sklearn.model_selection import train_test_split
```

```
tr_x, val_x, tr_y, val_y = train_test_split(X, y, test_size = 0.25,                                  random_state = 42)
```

```
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(tr_x, tr_y)

previsoes = model.predict(val_x)
```

```
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(val_y, previsoes)

print(f'A acurácia é {acuracia*100}%')
```
