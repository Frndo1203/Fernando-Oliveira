```
dado_ar1_p1 = dado_ar1[0:500] #treino
dado_ar1_p2 = dado_ar2[500:600]#teste
```

```
import statsmodels.tsa.api as smtsa
```

```
modelo_ar1 = smtsa.ARMA(dado_ar1_p1['Preco'], order = (1, 0)).fit() #ordem 1
```

```
modelo_previsto1 = modelo_ar1.predict(start = 500, end = 599)
```

```
plt.figure(figsize = (12, 6))
plt.plot(dado_ar1_p1['Data'], dado_ar1_p1['Preco'])
plt.plot(dado_ar1_p2['Data'], dado_ar1_p2['Preco'])
plt.plot(dado_ar1_p2['Data'], modelo_previsto1, 'r.')
```

```
dado_ar3_p1 = dado_ar3[0:500] #treino
dado_ar3_p2 = dado_ar3[500:600]#teste
```

```
modelo_ar3 = smtsa.ARMA(dado_ar3_p1['Preco'], order = (3, 0)).fit() #ordem 3
```

```
modelo_previsto3 = modelo_ar3.predict(start = 500, end = 599)
```

```
plt.figure(figsize = (12, 6))
plt.plot(dado_ar3_p1['Data'], dado_ar3_p1['Preco'])
plt.plot(dado_ar3_p2['Data'], dado_ar3_p2['Preco'])
plt.plot(dado_ar3_p2['Data'], modelo_previsto3, 'r.')
```

**(P,D,Q,m), inclui sazonalidade**

```
from pmdarima.arima.utils import nsdiffs
```

```
plt.plot(cresc['Data'], cresc['Preco'])
plt.xticks(rotation = 70);
```

```
D = nsdiffs(cresc_p1['Preco'].values, m = 2, max_D = 12,
        test = 'ch') #teste de Canova-Hansen
```

```
D
```

```
modelo_busca2 = auto_arima(cresc_p1['Preco'].values, start_p = 0,
                           start_q = 0, max_p = 6, max_q = 6,
                           d = 1, D = D, start_Q = 1, start_P = 1,
                           max_P = 4, max_Q = 4, m = 2,
                           seasonal = True, trace = True,
                           error_action = 'ignore', supress_warnings = True,
                           stepwise = False)
```

```
modelo_busca.aic()
```
